# 23-Reshaping a data frame

```python
fasla=pd.DataFrame([['12345',100,200,300],['34567',400,500,600],['67890',700,800,900]],
                       columns=['zip','factory','warehouse','retail'])
```

```python
fasla
```

```python
fasla2=pd.DataFrame([['1','12345','factory'],['2','34567','warehouse']],
                       columns=['user_id','zip','location_type'])
```

```python
fasla2
```

```python
fasla=fasla.melt(id_vars=['zip'], var_name='location_tyoe', value_name='distance')
fasla
```

```python
import seaborn as sns
sns.barplot(x='zip', y='distance', hue='location_tyoe', data=fasla)
```
